```python
def convert_to_tfidf_vectorizer(df, text_col, max_features=5000, ngram_range=(2,3)):
    df = df.copy()
    vectorizer = TfidfVectorizer(max_features=max_features, ngram_range=ngram_range)
    X_tfidf = vectorizer.fit_transform(df[text_col])
    
    return X_tfidf
```

```python
df = pd.read_pickle("df_cleaned_2.pkl")
```

```python
restaurant_reviews = pd.read_pickle("restaurant_reviews.pkl")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df[['Rating','reviews_preprocessed','Clean_Cuisine']]
```

```python
df.isnull().sum()
```

```python
X_reviews = convert_to_tfidf_vectorizer(df, "reviews_preprocessed", max_features=5000, ngram_range=(2,3))
```

```python
X_Cuisine = convert_to_tfidf_vectorizer(df, "Clean_Cuisine", max_features=10000, ngram_range=(3,4))
```

```python
scaler = StandardScaler()
```

```python
rating = df[['Rating']]
```

```python
rating_scaled = scaler.fit_transform(rating)
```

```python
df2 = hstack([X_reviews, X_Cuisine, rating_scaled])
```

```python
ssd = []
```

```python
for k in range(4, 10):
    model = KMeans(n_clusters=k, verbose=1, random_state=42)
    model.fit(df2)
    ssd.append(model.inertia_)
```

```python
plt.plot(range(4, 10), ssd, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (SSD)")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
scores = {}
for k in [7, 8, 9]:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(df2)
    score = silhouette_score(df2, labels, sample_size=20000, random_state=42)
    scores[k] = score
    print(f"k = {k}, silhouette score = {score:.4f}")
```

```python
kmeans = KMeans(n_clusters=8, random_state=101, n_init=10)
```

```python
kmeans.fit(df2)
```

```python
restaurant_reviews['cluster'] = kmeans.labels_
```

```python
def top_words_by_cluster(df, n_words=10):
    vectorizer = CountVectorizer(stop_words="english", max_features=5000)
    for cluster in sorted(df["cluster"].unique()):
        reviews = df.loc[df["cluster"] == cluster, "Reviews"].dropna()
        X = vectorizer.fit_transform(reviews)
        word_counts = X.sum(axis=0).A1
        vocab = vectorizer.get_feature_names_out()
        top_words = sorted(zip(vocab, word_counts), key=lambda x: x[1], reverse=True)[:n_words]
        print(f"\nCluster {cluster} top words: {top_words}")
```

```python
top_words_by_cluster(df3,n_words=20)
```
